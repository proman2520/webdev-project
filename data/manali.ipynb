```python
#importing necessary libraries 
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
#reading in csv file into a dataframe 
df_ufo = pd.read_csv('original.csv')
df_ufo
```

```python
#storing only the data from the USA into a new dataframe 
df_usa=df_ufo[df_ufo["country"]=="us"]
df_usa
```

```python
#dropping the "comments" and "duration" columns
df_scrubbed_usa=df_usa.drop("comments",axis=1)
df_scrubbed_usa=df_scrubbed_usa.drop("duration (hours/min)",axis=1)
df_scrubbed_usa
```

```python
df_scrubbed_usa = df_scrubbed_usa.rename(columns={'longitude ':'longitude'})
```

```python
# changing some rows in duration(seconds) so they are not strings
df_scrubbed_usa.at[27822,"duration (seconds)"]=2
df_scrubbed_usa.at[35692,"duration (seconds)"]=8

# changing some rows in datetime to make it in the right format to convert
df_scrubbed_usa.at[388,"datetime"]="10/12/2006 00:00"
df_scrubbed_usa.at[962,"datetime"]="10/2/2012 00:00"
df_scrubbed_usa.at[1221,"datetime"]="10/13/2012 00:00"
df_scrubbed_usa.at[1222,"datetime"]="10/13/2012 00:00"
df_scrubbed_usa.at[1317,"datetime"]="10/14/2004 00:00"
df_scrubbed_usa.at[1359,"datetime"]="10/14/2007 00:00"
df_scrubbed_usa.at[1445,"datetime"]="10/14/2012 00:00"
df_scrubbed_usa.at[1663,"datetime"]="10/15/2011 00:00"
```

```python
#changing all 24:00 to 23:59 so it is in the proper format to change into datetime type
df_scrubbed_usa["datetime"]=df_scrubbed_usa["datetime"].replace("24:00","23:59", regex=True)
```

```python
#changing the data types of the columns to make easier when grouping
df_scrubbed_usa["latitude"]=df_scrubbed_usa["latitude"].astype("float64")
df_scrubbed_usa["duration (seconds)"]=df_scrubbed_usa["duration (seconds)"].astype("float64")
df_scrubbed_usa["date posted"]=df_scrubbed_usa["date posted"].astype("datetime64")
df_scrubbed_usa["datetime"]=df_scrubbed_usa["datetime"].astype("datetime64")
```

```python
#creating an ID column to help with referencing
df_scrubbed_usa
ids = range(1, len(df_scrubbed_usa) + 1)
df_scrubbed_usa['ID'] = ids
df_scrubbed_usa = df_scrubbed_usa[['ID', 'datetime', 'city', 'state', 'country',\
                                  'shape', 'duration (seconds)', 'latitude', \
                                  'longitude', 'date posted']]
df_scrubbed_usa.head()
```

# Final

```python
# Export the data as a CSV so it has an ID column
# Rename the column to be consistent with SQL syntax for loading the data
df_scrubbed_usa = df_scrubbed_usa.rename(columns={"duration (seconds)":"duration_seconds", "date posted":"date_posted"})
display(df_scrubbed_usa.head(10))
df_scrubbed_usa.to_csv("final_data.csv", index=False)
```

```python
duration_ufo = df_scrubbed_usa['duration_seconds'].value_counts()
id_ufo = df_scrubbed_usa.groupby('duration_seconds')['ID'].count()

plt.style.use('dark_background')
plt.scatter(id_ufo, duration_ufo, color="lightgreen", edgecolor="black", alpha=0.75)
plt.xlabel('ID #')
plt.ylabel('Duration (sec)')
plt.title('How long each UFO was seen for')

plt.show()
```

```python
#getting the average time each state saw the UFO for
summary_agg = df_scrubbed_usa.groupby(['state'])[['duration_seconds']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_agg
```

```python
mean_state=df_scrubbed_usa['duration_seconds'].groupby(df_scrubbed_usa['state']).mean()

plt.style.use('dark_background')
state_avg= mean_state.plot.bar(color='lightgreen')
plt.xlabel('State')
plt.ylabel('Avg time (sec)')
plt.title('Average time per State')
plt.show()
```
